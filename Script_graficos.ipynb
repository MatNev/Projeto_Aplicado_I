```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel (r"C:\Users\joaov\Documents\Domiciliosatendidos.xlsx")
```

```python
# Cria um dataFrame com a somatória de domicílios por ano
domicilios_por_ano = df.groupby("Ano")["Domicilios"].sum()

# Cria um gráfico de barras para mostrar densidade
plt.title("Quantidade de domicílios afetados pelo programa Luz Para Todos")
plt.bar(domicilios_por_ano.index, domicilios_por_ano.values)
plt.xlabel("Ano")
plt.ylabel("Quantidade de domicílios")

plt.show()
```

```python
# Calcula a média de cada ano
media_por_ano = df.groupby("Ano")["Domicilios"].mean()

# Cria um gráfico de linhas
plt.plot(media_por_ano.index, media_por_ano.values, label="Média por ano")

# Adiciona títulos e rótulos aos eixos
plt.title("Média da quantidade de domicílios afetados por ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade de domicílios")
plt.legend()

# Calcula a média geral
media_geral = df["Domicilios"].mean()

# Adiciona a linha de média geral
plt.axhline(y=media_geral, color="black", linestyle="--", label="Média geral")
plt.legend()

# Exibe o gráfico
plt.show()
```

```python
# Cria um DataFrame com a somatória dos domicílios atendidos por estado
df_por_estado = df.groupby("Estado")["Domicilios"].sum()

# Cria um gráfico de barras para mostrar densidade
plt.title("Quantidade de domicílios por estado")
plt.bar(df_por_estado.index, df_por_estado.values)
plt.xticks(df_por_estado.index, rotation=90)
plt.xlabel("Estado")
plt.ylabel("Quantidade de domicílios")
plt.show()
```

```python
# Cria um gráfico de linhas mostrando a densidade da variável Ano
sns.displot(df["Ano"], kind="kde")

# Adiciona títulos e rótulos aos eixos
plt.title("Atuação ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Densidade de atuação")

# exibe o gráfico
plt.show()
```
